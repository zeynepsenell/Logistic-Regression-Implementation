```python
import numpy as np
import matplotlib.pyplot as plt
 
def create_data():
    #for creating 100 datas between -1 and 1. 
    X_array = np.array([])
    Y_array = np.array([])
    for i in range (100):
        x = np.random.uniform(low=-1, high=1)
        y = np.random.uniform(low=-1, high=1)
        X_array = np.append(X_array, x)
        Y_array = np.append(Y_array, y)
        
    return X_array,Y_array
 
def main():
    X,Y = create_data()
    #with learning rate of 0.1
    calculated_weight, calculated_bias = gradient_descent(X, Y, 0.1)
    #with learning rate of 0.01
    calculated_weight, calculated_bias = gradient_descent(X, Y, 0.01)
    #with learning rate of 0.001
    calculated_weight, calculated_bias = gradient_descent(X, Y, 0.001)
    
    #calculated predicted y values with last bias and weight
    Y_pred = calculated_weight*X + calculated_bias
     
    #for drawing values
    plt.figure(figsize = (6,4))
    plt.scatter(X, Y, marker='o', color='yellow')
    plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)])
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.show()
    
def gradient_descent(x, actual_y_value, learning_rate):
    current_weight = 0
    current_bias = 0
    learning_rate = learning_rate
    costs_in_100_step = []
    step = []
    costs = []
    weights = []
     
    for i in range(2000):
        predicted_y_value = (current_weight * x) + current_bias
        current_cost = np.sum((actual_y_value-predicted_y_value)**2) / len(actual_y_value)
        costs.append(current_cost)
        weights.append(current_weight)
        
        #for every 100 steps in 2000 iterations
        for j in range(20):
            if i == 100 * j:
                costs_in_100_step.append(current_cost)
                step.append(j)

        weight_d = sum(x * (actual_y_value-predicted_y_value)) * -(2/100) 
        bias_d = sum(actual_y_value-predicted_y_value) * -(2/100)
         
        current_weight = current_weight - (learning_rate * weight_d)
        current_bias = current_bias - (learning_rate * bias_d)
        #print(f"Iteration {i+1}: Cost {current_cost}, Weight {current_weight}")
     
    #draw_plot(weights, costs,"Weight","Cost",f"With learning rate {learning_rate}")
    draw_plot(step, costs_in_100_step,"Steps","Cost",f"With learning rate {learning_rate}")
    
    return current_weight, current_bias
 
def draw_plot(x_values, y_values, x_label, y_label, label ):
    plt.figure(figsize = (6,4))
    plt.plot(x_values, y_values)
    plt.scatter(x_values, y_values, marker='o', color='blue')
    plt.ylabel(y_label)
    plt.xlabel(x_label)
    plt.title(label)
    plt.show()
    

main()
```
